```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
N = 10
```

```python
pg_episode_rewards = np.load('pg_episode_rewards.npy')
pg_recent_avg_rewards = []
for i in range(pg_episode_rewards.shape[0]):
    recent_rewards = pg_episode_rewards[max(0, i - N):i + 1]
    pg_recent_avg_rewards.append(sum(recent_rewards) / len(recent_rewards))
```

```python
plt.clf()
plt.figure(figsize=(16, 5))
plt.title('Reward v.s. Episode')
plt.ylabel('Average reward in last %d episodes' % (N))
plt.xlabel('# of episodes')
plt.plot(pg_recent_avg_rewards)
plt.savefig('reward_episode_pg.png')
plt.show()
```
